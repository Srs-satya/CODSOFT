Preprocess the Data

```python
encoder = LabelEncoder()
new_col = encoder.fit_transform(df_combine["merchant"].values)
df_combine["merchant_new"] = new_col
df_combine.drop(labels = ["merchant"], axis = 1, inplace = True)
```

```python
encoder = LabelEncoder()
new_col1 = encoder.fit_transform(df_combine["category"].values)
df_combine["category_new"] = new_col1
df_combine.drop(labels = ["category"], axis = 1, inplace = True)
```

```python
df_combine = pd.get_dummies(df_combine)
df_combine.drop(labels=['gender_F'], axis = 1, inplace = True)
```

```python
df_combine.head()
```

```python
X = df_combine.drop("is_fraud", axis=True)
y = df_combine["is_fraud"]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, train_size=0.3, random_state = 42)
```

Training Logistic Model

```python
lr= LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
print("Logistic Regression Model: ")
print(confusion_matrix(y_test, lr_pred))
print(classification_report(y_test, lr_pred))
print("Accuracy:", accuracy_score(y_test, lr_pred))
```

Train DecisionTree Model

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
dt_pred = dt.predict(X_test)
```

```python
print(" Decision Tree Classifier: ")
print(confusion_matrix(y_test, dt_pred))
print(classification_report(y_test, dt_pred))
print("Accuracy:", accuracy_score(y_test, dt_pred))
```
